```python
import pandas as pd
import numpy as np
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer,KNNImputer
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.ensemble import BaggingRegressor
```

```python
df = pd.read_csv('./updated_datasets_distrubutions/nan_ds.csv')
df = df.drop('day',axis=1)
df_itr = df.copy()
df_knn = df.copy()
df_svr = df.copy()
df_bgr = df.copy()
```

```python
iterative_imp = IterativeImputer(HistGradientBoostingRegressor(),missing_values=np.nan,max_iter=1000)
```

```python
df_itr = iterative_imp.fit_transform(df_itr)
```

```python
df_itr = pd.DataFrame(data=df_itr,columns=df.columns)
```

```python
df_itr
```

```python
df_itr.to_csv('./updated_datasets_distrubutions/final_ds_itr.csv')
```

```python
knn_imp = KNNImputer(missing_values=np.nan,n_neighbors=10000,weights='uniform')
```

```python
df_knn = knn_imp.fit_transform(df_knn)
```

```python
df_knn = pd.DataFrame(data=df_knn,columns=df.columns)
```

```python
df_knn
```

```python
df_svr
```

```python
df_knn.to_csv('./updated_datasets_distrubutions/final_ds_knn.csv')
```

```python
svr_imp = IterativeImputer(SVR(),missing_values=np.nan,max_iter=1000)
df_svr = svr_imp.fit_transform(df_svr)
df_svr = pd.DataFrame(data=df_svr,columns=df.columns)
df_svr.to_csv('./updated_datasets_distrubutions/final_ds_svr.csv')
```

```python
bgr_imp = IterativeImputer(BaggingRegressor(),missing_values=np.nan,max_iter=1000)
df_bgr = bgr_imp.fit_transform(df_bgr)
df_bgr = pd.DataFrame(data=df_bgr,columns=df.columns)
df_bgr.to_csv('./updated_datasets_distrubutions/final_ds_bgr.csv')
```
