```python
import numpy as np
import pandas as pd
```

```python
file_paths = [
    './updated_datasets_distrubutions/final_ds_bgr.csv',
    './updated_datasets_distrubutions/final_ds_itr.csv',
    './updated_datasets_distrubutions/final_ds_knn.csv',
    './updated_datasets_distrubutions/final_ds_svr.csv',
]
```

```python
dfs = []
for path in file_paths:
    dfs.append(pd.read_csv(path))
```

```python
list(dfs[0].columns)
```

```python
unwanted_col = set(['sampleCount','noDataCount'])
```

```python
unwanted_col_names = [str(col_name) for col_name in list(dfs[0].columns) for unwanted in unwanted_col if unwanted in col_name] + ['Unnamed: 0']
```

```python
for i,df in enumerate(dfs):
    dfs[i] = df.drop(unwanted_col_names,axis=1)
```

```python
for file_name,df in zip(file_paths,dfs):
    df.to_csv(file_name,index=False)
```
